```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
riqueza_comuna = pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv')
poblacion = pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/pobXedadCenso.csv')
vacunacion = pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto81/vacunacion_comuna_edad_1eraDosis.csv')
```

```python
poblacion = poblacion.replace('*',0)
poblacion.head()
```

```python
vacunacion.head()
```

```python
vacunacion['TOTAL_VAC'] = vacunacion.iloc[:, 5:71].sum(axis=1)
vacunacion = vacunacion.loc[:,['Comuna','Codigo comuna','TOTAL_VAC']] 
```

```python
vacunacion[(vacunacion['Comuna'] == 'Alhue')]
```

```python
poblacion.dtypes
```

```python
poblacion['EDAD_0A5']=poblacion['EDAD_0A5'].astype(str).astype(int)
poblacion['EDAD_6A14']=poblacion['EDAD_6A14'].astype(str).astype(int)
poblacion['EDAD_15A64']=poblacion['EDAD_15A64'].astype(str).astype(int)
poblacion['EDAD_65YMAS']=poblacion['EDAD_65YMAS'].astype(str).astype(int)
poblacion.dtypes
```

```python
poblacion = poblacion.groupby('COMUNA').sum().reset_index()
poblacion = poblacion.rename(columns={'COMUNA':'CODIGO_COMUNA'})
poblacion
```

```python
poblacion['TOTAL'] = poblacion.iloc[:, [1,2,3,4]].sum(axis=1)
poblacion.head()
```

```python
vacunacion.dtypes
```

```python
vacunacion = vacunacion.rename(columns={'Comuna':'NOMBRE_COMUNA','Codigo comuna':'CODIGO_COMUNA', 'Poblacion':'POBLACION'})
```

```python
join1 = pd.merge(vacunacion, poblacion, on='CODIGO_COMUNA')
join1
```

```python
join1[(join1['NOMBRE_COMUNA'] == 'San Pedro')]
```

```python
riqueza_comuna = riqueza_comuna.rename(columns={'comuna':'CODIGO_COMUNA'})
```

```python
df = pd.merge(join1, riqueza_comuna, on='CODIGO_COMUNA')
df
```

```python
df['PORCENTAJE_VAC']=df['TOTAL_VAC']/df['TOTAL']
df['PORCENTAJE_VAC_65+']=df['EDAD_65YMAS']/df['TOTAL']
df
```
